```python
import matplotlib.pyplot as plt
value=[23,45,12,6,34]
year=[2001,2002,2003,2004,2005]
plt.plot(year,value)
plt.scatter(year,value,s=value)
plt.title('world population')
plt.ylabel('no of poplution per acre')
plt.xlabel('year')
plt.yticks([12,34,55,66],['p1','p2','p3','p4'])
plt.xticks([2000,2002,2004,2006],['y1','y2','y3','y4'])
plt.grid(True)
plt.show()
```
