## Linear Regression using Single Variable

```python
import pandas as pd
from sklearn import linear_model
from matplotlib import pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('Salaries.csv')
df.head()
```

```python
df.columns
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['service']],df.salary)
```

```python
predicted_salary=reg.predict(df[['service']])
```

```python
x=plt.xlabel('service')
y=plt.ylabel('salary')
plt.scatter(df['service'],predicted_salary,color='red',marker='*',label='pointer')
plt.plot(df['service'],predicted_salary,marker='+',color='black')
plt.show()
```

```python
x=plt.xlabel('service')
y=plt.ylabel('salary')
plt.scatter(df['service'],df['salary'],marker='+',color='blue')
plt.show()
```

```python
x=plt.xlabel('service')
y=plt.ylabel('salary')
plt.plot(df.service,df.salary,color='violet',label='line1')
plt.scatter(df['service'],df['salary'],color='black',marker='*',label='pointer')
plt.legend()
plt.show()
```
