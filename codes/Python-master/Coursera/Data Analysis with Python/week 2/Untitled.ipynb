# Binning

![Binning](Binning.png)

```python
import pandas as pd
df=pd.read_csv('Price.csv')
df
```

```python
import numpy as np
bins=np.linspace(min(df['Price']),max(df['Price']),4)
bins
```

```python
group_names=['Min','Mid','High']
```

```python
df['Price_binned']=pd.cut(df['Price'],bins,labels=group_names,include_lowest=True)
df
```
