# MSE(Mean Square Error)
$$
mse=\frac{1}{n} \sum_{i=1}^n (y_i-y_p)^2\\
mse=\frac{1}{n} \sum_{i=1}^n(y_i-(mx_i+b))^2
$$
mse=cost function<br>
p=predict

![](gr-pic.png)

![](eq_idea.png)

$$
f(x,y)=x^3+y^2 \\
\frac{\partial f}{\partial x}=3x^2+0=3x^2 \\
\frac{\partial f}{\partial y}=0+2y=2y
$$

![](equation.png)

$$
mse=\frac{1}{n} \sum_{i=1}^n(y_i-(mx_i+b))^2\\
\frac{\partial}{\partial m}=\frac{2}{n}\sum_{i=1}^n-x_i(y_i-(mx_i+b))\\
\frac{\partial}{\partial b}=\frac{2}{n}\sum_{i=1}^n-(y_i-(mx_i+b))
$$

### learning rate
$$
m=m-learningrate*\frac{\partial}{\partial m}\\
b=b-learningrate*\frac{\partial}{\partial b}
$$

```python
import matplotlib.pyplot as plt
import numpy as np
def greadient(x,y):
    m=b=0
    n=len(x)
    iterations=1000
    learningrate=0.08
    for i in range(iterations):
        y_predict=m*x+b
        plt.plot(x,y_predict)
        cost=(1/n)*sum(val**2 for val in (y-y_predict))
        md=(-2/n)*sum(x*(y-y_predict))
        bd=(-2/n)*sum(y-y_predict)
        m=m-learningrate*md
        b=b-learningrate*bd
        print('m {}, b {}, cost {}, iterations {}'.format(m,b,cost,iterations))
    plt.show()
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
greadient(x,y)
```
