```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('homeprices.csv')
```

```python
df
```

# method - 1 using dummy variable

```python
dummy=pd.get_dummies(df.town)
dummy
```

```python
merged=pd.concat([df,dummy],axis=1)#or axis='columns'
merged
```

```python
# drop any one dummy  varible always
final=merged.drop(['town','west windsor'],axis='columns')
final
```

```python
model=linear_model.LinearRegression()
```

```python
X=final.drop(['price'],axis=1)
X
```

```python
y=df.price
y
```

```python
model.fit(X,y)
```

```python
model.predict([[2900,0,1]])
```

```python
model.score(X,y)#check how much our model is accurate here our model 95% accurate
```
