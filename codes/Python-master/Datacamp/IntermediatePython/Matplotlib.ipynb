# Matplotlib

```python
import matplotlib.pyplot as plt
year=[2000,2001,2002,2006,2014,2020]
pop=[1,4,6,2,7,3]
plt.plot(year,pop)
plt.show()
```

```python
import matplotlib.pyplot as plt
year=[2000,2001,2002,2006,2014,2020]
pop=[1,4,6,2,7,3]

plt.scatter(year,pop)
# plt.grid(True)
plt.show()
```

```python
year=[2000,2001,2002,2006,2014,2020]
pop=[1,4,6,2,7,3]
plt.plot(year,pop)

plt.yscale('log')
plt.show()
```

```python
year=[2000,2001,2002,2006,2014,2020]
pop=[1,4,6,2,7,3]
plt.plot(year,pop)

plt.xscale('log')
plt.show()
```

### Customization 

```python
year=[2000,2001,2002,2006,2014,2020]
pop=[1,4,6,2,7,3]
plt.plot(year,pop)

plt.xlabel('Year')
plt.ylabel('pop')
plt.title('World Population Projects')
plt.show()
```

### Ticks

```python
plt.yticks([1,6,4,2,3,5,7])
plt.show()
```

```python
plt.yticks([1,4,7,3,5,8],['a','b','c','d','e','f'])
plt.show()
```

```python
plt.yticks([1,4,7,3,5,8],['a','b','c','d','e','f'])
plt.grid(True)
plt.show()
```

### Sizes

```python
import numpy as np
year=[2000,2001,2002,2006,2014,2020]
pop=[1,4,6,2,7,3]

# convert in numpy array
np_pop=np.array(pop)
np_pop*=40
plt.scatter(year,pop,s=np_pop)

plt.show()
```

### Color

```python
import numpy as np
year=[2000,2001,2002,2006,2014,2020]
pop=[1,4,6,2,7,3]

# convert in numpy array
np_pop=np.array(pop)
np_pop*=40
plt.scatter(c='green',x=year,y=pop,s=np_pop,alpha=0.8)

plt.show()
```

### Additional customization

#### text

```python
import matplotlib.pyplot as plt
import numpy as np
year=[2000,2001,2002,2006,2014,2020]
pop=[1,4,6,2,7,3]

# convert in numpy array
np_pop=np.array(pop)
np_pop*=40
plt.scatter(c='green',x=year,y=pop,s=np_pop,alpha=0.8)

# addtional customization
plt.text(2000,1,'India')
plt.text(2014,7,'KKKKK')

plt.grid(True)
plt.show()
```
